### Cleaning Data

Membersihkan dataframe pada tabel Hour

```python
#Mendefinisikan variabel date_time_colum yang memuat array "dteday"
date_time_column = ["dteday"]

#Melakukan perulangan for bila variabel column berada di dalam variabel date_time_column
for column in date_time_column :
  #Melakukan perubahan tipe data khusus pada kolom dteday pada tabel hour
  hour_df[column] = pd.to_datetime(hour_df[column])
```

```python
#Memastikan bahwa tipe data telah berubah
hour_df.info()
```

Membersihkan dataframe pada tabel Day

```python
# Mendefinisikan variabel date_time_colum yang memuat array "dteday"
date_time_column = ["dteday"]

#Melakukan perulangan for bila variabel column berada di dalam variabel date_time_column
for column in date_time_column :
  #Melakukan perubahan tipe data khusus pada kolom dteday pada tabel hour
  day_df[column] = pd.to_datetime(day_df[column])
```

```python
#Memastikan tipe data telah berubah
day_df.info()
```

## Exploring Data Analysis (EDA)

```python
#Menampilkan ringkasan statistik tabel hour dari data numerik dalam DataFrame secara keseluruhan
hour_df.describe(include="all")
```

```python
#Menampilkan ringkasan statistik tabel day dari data numerik dalam DataFrame secara keseluruhan
day_df.describe(include="all")
```

```python
# Membuat kamus weathersit_labels untuk memetakan kode cuaca (dalam kolom 'weathersit_day') ke label cuaca yang sesuai.
# Menggunakan metode .map()
weathersit_labels = {
    1: 'Cerah',
    2: 'Berawan',
    3: 'Hujan'
}
day_df['weathersit_label'] = day_df['weathersit'].map(weathersit_label)
```

```python
#Melakukan pengelompokkan pada kolom weathersit_label yang telah dibuat
#Kemudian melakukan perhitungan rata-rata jumlah sepeda yang disewa menurut cuaca
#Kemudian bisa digunakan sort_values('cnt') untuk mengurutkan jumlah sepeda yang disewa
day_df.groupby('weathersit_label')['cnt'].mean().reset_index().sort_values('cnt')
```

```python
#Melakukan pengelompokkan kolom working day dan cnt, lalu
#Menghitung rata-rata pada 'cnt' sesuai kolom workingday
day_df.groupby('workingday')['cnt'].mean().reset_index().sort_values('cnt')
```

```python
#Melakukan pendefinisian label
#dengan membuat kamus pada kolom weekday dan mengisinya sesuai nama hari
weekday_labels = {
    1: "Senin",
    2: "Selasa",
    3: "Rabu",
    4: "Kamis",
    5: "Jumat",
    6: "Sabtu",
    7: "Minggu"
}
#Memasukkan label yang telah dibuat pada kolom weekday dengan fungsi .map
day_df['weekday_label'] = day_df['weekday'].map(weekday_labels)
```

```python
day_df.groupby('weekday_label')['cnt'].mean().reset_index()
```

```python
#Mendefinisikan total sewa yang terdaftar pada kolom registered dengan sum()
total_terdaftar = day_df['registered'].sum()
#Mendefinisikan total sewa yang tidak terdaftar pada kolom casual dengan sum()
total_tidak_terdaftar = day_df['casual'].sum()

#Menampilkan jumlah sewa yang terdaftar atau tidak dengan pemanggilan masing-masing variabel
print("Jumlah sewa sepeda yang terdaftar adalah :", total_terdaftar)
print("Jumlah sewa sepeda yang tidak terdaftar adalah :", total_tidak_terdaftar)
```

```python
#Melakukan pendefinisian label
#dengan membuat kamus pada kolom season dan mengisinya sesuai musim
season_labels = {
    1: "Dingin",
    2: "Semi",
    3: "Panas",
    4: "Gugur"
}
#Memasukkan label yang telah dibuat pada kolom season dengan fungsi .map
day_df['season_label'] = day_df['season'].map(season_labels)
```

```python
#Melakukan pengabungan kolom season_label dan rata-rata kolom cnt
#lalu melakukan pengaturan ulang dataframe dengan reset_index
#lalu mengurutkan jumlah dengan sort_values
day_df.groupby('season_label')['cnt'].mean().reset_index().sort_values('cnt')
```

## Visualisasi Data

```python
#Mendefinisikan variabel avg_workday yang diambil dari rata-rata jumlah pada kolom workingday
avg_workday = day_df.groupby('workingday')['cnt'].mean()
#Mendefinisikan variabel avg_weekday yang diambil dari rata-rata jumlah pada kolom weekday
avg_weekday = day_df.groupby('weekday')['cnt'].mean()

#Memberi label pada chart workday dengan nilai Hari Kerja dan Hari Libur
label_workday = ['Hari Kerja', 'Hari Libur']
#Memberi label pada chart weekday dengan nilai hari dari senin sampai minggu
label_weekday = ['Senin','Selasa','Rabu','Kamis','Jumat','Sabtu','Minggu']

# Plot
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Pie chart untuk rata-rata pada hari kerja
axes[0].pie(avg_workday, labels=label_workday, autopct='%1.1f%%', colors=('#265073', '#9AD0C2'))
axes[0].set_title('Rata-rata Penyewaan Sepeda pada Hari Kerja')

# Pie chart untuk rata-rata pada setiap hari
axes[1].pie(avg_weekday, labels=label_weekday, autopct='%1.1f%%', colors=('#2D9596', 'lightgrey', '#9AD0C2', '#265073'))
axes[1].set_title('Rata-rata Penyewaan Sepeda pada Hari dalam Seminggu')

# Menampilkan plot
plt.show()
```

```python
#Mendefinisikan rata-rata jumlah pengguna sepeda berdasarkan cuaca
avg_weather = day_df.groupby('weathersit_label')['cnt'].mean().reset_index()
plt.figure(figsize=(10,7))
#Ambil nilai x dan y dari weathershit_label dan cnt diikuti pewarnaan bar pada chart
plt.bar(avg_weather['weathersit_label'], avg_weather['cnt'], color="#2D9596")
#Menambahkan judul
plt.title('Rata-rata jumlah pengguna sepeda berdasarkan kondisi cuaca')
#Menambahkan keterangan pada garis x
plt.xlabel('Kondisi cuaca')
#Menambahkan keterangan pada garis y
plt.ylabel('Rata-rata jumlah pengguna sepeda')
# Memutar label sumbu x sebesar 45 derajat
plt.xticks(rotation=45)

# Menambahkan grid pada sumbu y
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan plot
plt.show()
```

```python
#Mendefinisikan variabel avg_season untuk mengitung rata-rata jumlah
#pengguna sepeda sesuai season atau musim
avg_season = day_df.groupby('season_label')['cnt'].mean().reset_index()
plt.figure(figsize=(10,7))
#Ambil nilai x dan y dari season_label dan cnt diikuti pewarnaan bar pada chart
plt.bar(avg_season['season_label'], avg_season['cnt'], color="#2D9596")
#Judul
plt.title('Rata-rata pengguna sepeda pada 4 musim')
#Menambahkan keterangan pada garis x
plt.xlabel('Musim')
#Menambahkan keterangan pada garis y
plt.ylabel('Rata-rata pengguna sepeda')
# Memutar label sumbu x sebesar 45 derajat
plt.xticks(rotation=45)

# Menambahkan grid pada sumbu y
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Mendefinisikan variabek count_bikesharing, yakni jumlah dari kolom registered dan casua;
count_bikesharing = [day_df['registered'].sum(), day_df['casual'].sum()]
#Mendefinisikan label
labels = ['Terdaftar', 'Tidak Terdaftar']
plt.figure(figsize=(8, 6))
#Memanggil variabel count_bikesharing, label untuk membuat bar chart
plt.barh(labels, count_bikesharing, color=['#2D9596', '#9AD0C2'])
#judul
plt.title('Jumlah Sewa Sepeda yang Terdaftar ataupun Tidak Terdaftar')
#keterangan pada garis x
plt.xlabel('Jumlah Sewa')
#keterangan pada garis y
plt.ylabel('Status')
#Menampilkan chart
plt.show()
```
